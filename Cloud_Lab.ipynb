```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Create a small dataset
data = {
    "Name": ["Alice", "Bob", "Charlie", "Diana", "Ethan"],
    "StudyHours": [5, 3, 8, 2, 6],
    "FinalGrade": [85, 70, 90, 60, 88]
}

df = pd.DataFrame(data)

# Step 2: Display the dataset
print("Dataset:")
print(df)

# Step 3: Basic statistics
print("\nBasic statistics:")
print(df.describe())

# Step 4: Visualization - Study Hours vs Final Grade
plt.figure(figsize=(6,4))
sns.scatterplot(x="StudyHours", y="FinalGrade", data=df)
plt.title("Study Hours vs Final Grade")
plt.xlabel("Study Hours")
plt.ylabel("Final Grade")
plt.show()

# Step 5: Save results (Binder feature)
# This demonstrates saving files in Binder
df.to_csv("grades_analysis_results.csv", index=False)
print("\nResults saved as 'grades_analysis_results.csv'")
```
